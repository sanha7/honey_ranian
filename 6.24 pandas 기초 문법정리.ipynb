```python
df1 = pd.DataFrame({'이름':['A', 'B', 'D'],
                   '성별':['남자','여자','남자']})
df1 #데이터프레임 두개 생성할거임
```

```python
df2 = pd.DataFrame({'이름':['A', 'B', 'C'],
                   '키':[180, 160, 170]})
df2
```

```python
pd.merge(df1, df2) #두 데이터프레임을 하나로 합침(join)
```

```python
pd.merge(df1, df2, how ='left') #left조인_왼쪽은 다 포함해서 교집합 만들기
```

```python
pd.merge(df1, df2, how ='right') #right조인_오른쪽은 다 포함해서 교집합 만들기
```

```python
pd.merge(df1, df2, how='outer') #합집합으로
```

```python
pd.concat([df1, df2]) #동일한 column을 가진 두 dataframe을 합치는게 concat
```

```python
pd.concat([df1,df2], axis=1) #위아래 말고 column 기준으로, 옆으로 합치고 싳으면 axis=1
```
